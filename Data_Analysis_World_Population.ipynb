```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("world_population.csv")
```

```python
# Ensure that floating point numbers are displayed with 2 decimal places
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
# This function calculates the null values and their percentages in the data, 
# providing insights into the necessity of data imputation and how it can be done.
def null_value_info():
    global df_null_data
    total_nulls = df.isnull().sum()
    percentage_nulls = (total_nulls / len(df)) * 100;
    df_null_data = pd.DataFrame({"Total Null Count":total_nulls,
                                 "Percentage": percentage_nulls})
```

```python
null_value_info()
df_null_data.T
```

```python
# Selects columns from the DataFrame that contain numerical data types.
# This allows for focusing on columns that contain numeric values for further analysis or processing.
numeric_df = df.select_dtypes(include=["number"])

# Selects columns from the DataFrame that contain numerical data types.
# This allows for focusing on columns that contain numeric values for further analysis or processing.
df_corr = numeric_df.corr()

# reates a heatmap visualization of the correlation matrix.
# This helps to visually identify patterns and relationships between variables in the dataset.
plt.figure(figsize=(10, 8))
sns.heatmap(df_corr, annot = True, cmap = "magma", center = 0, linewidths = 0.5, linecolor = "white")
plt.title("Correlation Matrix")
plt.show()
```

```python
numeric_df['Continent'] = df['Continent']
numeric_df.groupby("Continent").mean().sort_values(by ="2022 Population", ascending = False)
```

```python
# Calculates the average population for each year column in the dataset, grouped by continent.
# This process involves taking the mean population value for each year within each continent,
# providing insights into the trend of population change over time across different continents.
df_sorted = numeric_df.groupby("Continent")[["1970 Population","1980 Population","1990 Population","2000 Population","2010 Population","2015 Population","2020 Population","2022 Population"]].mean().sort_values(by ="2022 Population", ascending = False)
```

```python
# Transposing the DataFrame to visualize the data in a more understandable table format.
# This process swaps the rows and columns of the DataFrame, making it easier to interpret
# the average population values for each year across continents.
df_transpose = df_sorted.transpose()
```

```python
df_transpose
```

## Population Increase by Continent Over the Years

```python
fig, ax = plt.subplots(figsize=(13, 5))
df_transpose.plot(ax=ax)
ax.set_title('Population Increase by Continent Over the Years', color = "#02485c", fontsize = 20)
plt.show()
```

## Comparing Top 10 Countries With The Highest And Lowest Growth Rates

```python
biggest_growth = df.sort_values(by = ["Growth Rate"], ascending = False).head(10)
slowest_growth = df.sort_values(by = ["Growth Rate"], ascending = True).head(10)
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(17, 13))
ax1.bar(biggest_growth["Country"],biggest_growth["Growth Rate"] ,color =sns.color_palette("viridis"))
ax1.set_title("Top 10 Biggest Growth Rate", color = "#02485c", fontsize = 20)
ax1.set_yticks([0,.2,.4,.6,.8,1,1.04,1.07])

ax2.bar(slowest_growth["Country"],slowest_growth["Growth Rate"], color = sns.color_palette("magma"))
ax2.set_title("Top 10 Slowest Growth Rate", color = "#02485c", fontsize = 20)
ax2.set_yticks([0,.2,.4,.6,.8,.85,.9,.95,1,1.05])

plt.subplots_adjust(hspace = 0.3)
plt.show()
```

## Most and Least Populous Countries by 2022

```python
most_crowded = df.sort_values(by = ["2022 Population"], ascending = False).head(10)
least_crowded = df.sort_values(by = ["2022 Population"], ascending = True).head(10)
```

```python
#In the Horizontal Bar Graphs, the first element (index = 0) is positioned at the bottom of the graph, closest to the x-axis. 
#For visual reasons, the most_crowded variable is sorted again so that the most crowded country, China, is in the 10th position (index = 9). 
#However, this is not a general sorting over the entire dataset.
most_crowded = most_crowded.sort_values(by = ["2022 Population"], ascending = True)
```

```python

fig, (ax1, ax2) = plt.subplots(2,1, figsize=(17, 8))
ax1.barh(most_crowded["Country"],most_crowded["2022 Population"] ,color =sns.color_palette("viridis",10))
ax1.set_title("Top 10 Most Crowded Country",color = "#02485c", fontsize = 20)

ax2.barh(least_crowded["Country"],least_crowded["2022 Population"], color = sns.color_palette("magma",10))
ax2.set_title("Top 10 Least Crowded Country", color = "#02485c", fontsize = 20)
ax2.set_xticks(np.arange(0,least_crowded["2022 Population"].iloc[9], step = least_crowded["2022 Population"].iloc[9] /8))
plt.subplots_adjust(hspace = 0.4)
plt.show()
```

```python
import plotly.express as px

fig = px.scatter_geo(most_crowded,
                     locations="Country",
                     locationmode="country names",
                     color="2022 Population",
                     hover_name="Country",
                     size="2022 Population",
                     projection="natural earth",
                     title="Most Populous Countries in 2022",
                     color_continuous_scale=px.colors.sequential.Plasma,
                     size_max=40)
fig.update_layout(
    title_text = "Most Populous Countries in 2022",
    title_x = 0.5,
    geo=dict(
        showframe = False,
        showcoastlines = True,
        projection_type = "natural earth"
    ),
     width=1200,
     height=600
)

fig.show()
```

## Most And Least Populus Countries In Every Continent

```python
most_populous_countries = df.loc[df.groupby('Continent')['2022 Population'].idxmax()]

plt.figure(figsize=(18,4))
plt.bar(most_populous_countries['Continent']+"/"+most_populous_countries['Country'], most_populous_countries['2022 Population'], color= sns.color_palette('viridis'))
plt.xlabel('Continent/City', color = "#468191", fontsize = 15)
plt.ylabel('Population', color = "#468191", fontsize = 15)
plt.title('Most Populous Contry in Each Continent (2022)', color = "#02485c", fontsize = 20)
plt.show()
```

```python
import plotly.express as px

fig = px.scatter_geo(
    most_populous_countries,
    locations="Country",
    locationmode="country names",
    hover_name="Country",
    size="2022 Population",
    size_max=50,
    projection="natural earth",
    title="Most Populous Country in Each Continent (2022)",
    color="Continent"
)

# Grafik düzenlemeleri
fig.update_layout(
    title_text='Most Populous Country in Each Continent (2022)',
    title_x=0.5,
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='equirectangular'
    ),
    width=1000,  # Set the width of the map
    height=600   # Set the height of the map
)

# Grafiği göster
fig.show()
```

```python
least_populous_countries = df.loc[df.groupby('Continent')['2022 Population'].idxmin()]

plt.figure(figsize=(18,4))
plt.bar(least_populous_countries['Continent']+"/"+least_populous_countries['Country'], least_populous_countries['2022 Population'], color= sns.color_palette('viridis'))
plt.xlabel('Continent/City', color = "#468191", fontsize = 15)
plt.ylabel('Population', color = "#468191", fontsize = 15)
plt.title('Least Populous Contry in Each Continent (2022)', color = "#02485c", fontsize = 20)
plt.show()
```
